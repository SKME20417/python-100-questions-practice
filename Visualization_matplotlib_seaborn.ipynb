```python
petal_length = df['petal_length']
petal_width = df['petal_width']
```

```python
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
viriginca = df[df['species'] == 'virginica']
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'], color = 'green', label = 'setosa')
plt.scatter(versicolor['petal_length'], versicolor['petal_width'], color = 'red', label = 'versicolor')
plt.scatter(viriginca['petal_length'], viriginca['petal_width'], color = 'yellow', label = 'viriginca')
plt.legend()
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```
